```python
import json
import pandas as pd
import glob

# Initialize a list to hold the data
data = []
gpt_models = ["claude"]
# Iterate over all JSON files in the specified directory
for file_path in glob.glob("result/**/*.json", recursive=True):
    with open(file_path, "r") as file:
        results = [json.loads(line) for line in file]

        # Extract model and category from file path
        path_parts = file_path.split('/')
        model = path_parts[-2]
        category = path_parts[-1]
        category = category.split('BFCL_v3_')[1].split('_result.json')[0]
        version = {
            "live": "v2",
            "multi_turn": "v3",
            "exec": "v1"
        }

        # derive version from category, if category string contains the key in version, use the value as version, otherwise use "v1"
        category_version = next((version[key] for key in version if key in category), "v1")
        print(category, category_version)
        if not any(model.startswith(m) for m in gpt_models):
            continue
        for entry in results:
            model_id = entry["id"]
            input_token_counts = entry.get("input_token_count", 0)
            output_token_counts = entry.get("output_token_count", 0)
            prompt_write_cache_token_counts = entry.get("prompt_write_cache_token_count", 0)
            prompt_read_cache_token_counts = entry.get("prompt_read_cache_token_count", 0)
            print(model, category)
            print("input_token_counts, output_token_counts", input_token_counts, output_token_counts)

            model_input_price = INPUT_PRICE_PER_MILLION_TOKEN[model] / 1000000
            model_output_price = OUTPUT_PRICE_PER_MILLION_TOKEN[model] / 1000000
            prompt_write_cache_price = PROMPT_WRITE_CACHE_PRICE_PER_MILLION_TOKENS[model] / 1000000
            prompt_read_cache_price = PROMPT_READ_CACHE_PRICE_PER_MILLION_TOKENS[model] / 1000000

            if isinstance(input_token_counts, int) and isinstance(output_token_counts, int):
                input_price = input_token_counts * model_input_price
                output_price = output_token_counts * model_output_price
                prompt_write_cache_price = prompt_write_cache_token_counts * prompt_write_cache_price
                prompt_read_cache_price = prompt_read_cache_token_counts * prompt_read_cache_price
                data.append(
                    {
                        "model": model,
                        "category_version": category_version    ,
                        "category": category,
                        "input_token_count": input_token_counts,
                        "output_token_count": output_token_counts,
                        "prompt_write_cache_token_count": prompt_write_cache_token_counts,
                        "prompt_read_cache_token_count": prompt_read_cache_token_counts,
                        "input_price": input_price,
                        "output_price": output_price,
                        "prompt_write_cache_price": prompt_write_cache_price,
                        "prompt_read_cache_price": prompt_read_cache_price,
                    }
                )
            else:
                # Sum up all the numbers in the lists of lists
                total_input_count = sum(sum(sublist) for sublist in input_token_counts)
                total_output_count = sum(sum(sublist) for sublist in output_token_counts)
                total_prompt_write_cache_count = sum(sum(sublist) for sublist in prompt_write_cache_token_counts)
                total_prompt_read_cache_count = sum(sum(sublist) for sublist in prompt_read_cache_token_counts)
                print("total_input_count, total_output_count", total_input_count, total_output_count)

                input_price = total_input_count * model_input_price
                output_price = total_output_count * model_output_price
                prompt_write_cache_price = total_prompt_write_cache_count * prompt_write_cache_price
                prompt_read_cache_price = total_prompt_read_cache_count * prompt_read_cache_price
                data.append(
                    {
                        "model": model,
                        "category_version": category_version,
                        "category": category,
                        "input_token_count": total_input_count,
                        "output_token_count": total_output_count,
                        "prompt_write_cache_token_count": total_prompt_write_cache_count,
                        "prompt_read_cache_token_count": total_prompt_read_cache_count,
                        "input_price": input_price,
                        "output_price": output_price,
                        "prompt_write_cache_price": prompt_write_cache_price,
                        "prompt_read_cache_price": prompt_read_cache_price,
                    }
                )

# Create a DataFrame from the collected data
df = pd.DataFrame(data)

# Display the DataFrame
df
```
