```python
import pandas as pd

# Path to the downloaded Excel file
file_path = "/content/tool_calling_table - Sheet2.csv"  # Replace with the actual file path

# Load the Excel file
df = pd.read_csv(file_path, skiprows=3)

# Display the DataFrame
df
```

```python
# Filter out rows where "License" contains "Proprietary"
df_filtered = df[~df["License"].str.contains("Proprietary", na=False)]
df_filtered
```

```python
headers = df_filtered.columns.tolist()
headers
```

```python
accuracy_features = [headers[1]]+headers[4:-2]  # Picking specific indices
accuracy_features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

x_axis = df_filtered["size"]
for col in accuracy_features:
  print(col)
  plt.figure(figsize=(4,3))
  print(df_filtered[col].shape, df_filtered["size"].shape)
  sns.scatterplot(x=x_axis, y=df_filtered[col])
  plt.title(f"Model Size vs {col}")
  plt.xlabel("Model Size")
  plt.ylabel(col)
  plt.grid(True)
  plt.show()
```

```python
df_filtered2 = df[df["size"]<200]
x_axis = df_filtered2["size"]
for col in accuracy_features:
  plt.figure(figsize=(4,3))
  sns.scatterplot(x=x_axis, y=df_filtered2[col])
  plt.title(f"Model Size vs {col}")
  plt.xlabel("Model Size")
  plt.ylabel(col)
  plt.grid(True)
  plt.show()
```

```python
df_numerical = df_filtered[["size"] + accuracy_features]#.apply(pd.to_numeric, errors="coerce")

size_correlation = df_numerical.corr()

# Print the correlation values
print("Correlation of Model Size with Accuracy Features:")
print(size_correlation.loc["size"])

# Plot heatmap for better visualization
plt.figure(figsize=(20, 12))
sns.heatmap(size_correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation of Model Size with Accuracy Features")
plt.show()
```
